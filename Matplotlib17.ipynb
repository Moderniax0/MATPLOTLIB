```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as seaborn
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

Predefined Styles

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

Membuat Style Sendiri
Saya membuat style dengan konfigurasi: axes.titlesize : 25 axes.labelsize : 20 lines.linewidth : 3 lines.markersize : 12 xtick.labelsize : 12 ytick.labelsize : 12

```python
plt.style.use('newStyle.mplstyle')
```

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Plot')

plt.show()
```

Menerapkan multiple style

```python
# Set Seaborn style
seaborn.set_style('whitegrid')  # Choose any Seaborn style you prefer

# Use the 'dark_background' style from Matplotlib
plt.style.use('dark_background')
```

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

Menerapkan temporary styling

```python
plt.style.use('ggplot')

plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```

```python
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')

plt.show()
```
